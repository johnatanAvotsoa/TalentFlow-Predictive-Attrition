```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv('../dataset/cleaned_data.csv')
df.head()
```

# DATA ANALYSIS

### EVALUATION PER DEPARTMENT 

```python
department_satisfaction_rank = df.groupby('Department')['satisfaction_level'].mean().sort_values(ascending=False).reset_index()
sns.set_theme(style='whitegrid')
plt.figure(figsize=(16,4))
sns.barplot(
    data=department_satisfaction_rank,
    x = 'Department', 
    y = 'satisfaction_level'
)
plt.title("Department average satisfaction (ordered)",fontsize=15,fontweight='bold')
plt.ylim(0,1)
plt.show()
```

```python
left_by_dep = df.loc[df['left']==1,'Department'].value_counts()
plt.figure(figsize=(20,6))
plt.pie(
    left_by_dep, 
    labels= left_by_dep.index,
    startangle=140 ,
    autopct='%1.1f%%',
    colors = sns.color_palette("pastel"),
    explode=[0.05] * len(left_by_dep)
)
centre_circle = plt.Circle((0,0), 0.30, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title('Volume of resignation by Department, implying recruitment cost', fontsize=18,fontweight='bold',pad=50)
plt.axis('equal')
plt.show()
```
